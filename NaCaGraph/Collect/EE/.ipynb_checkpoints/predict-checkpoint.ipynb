```python
import os
import glob
from utils import read_json
from utils import sentencize
from utils import build_input_for_ee
from utils import listdir_nohidden
from argparse import ArgumentParser
import dill as pickle
import time

import spacy
import torch
from OmniEvent.infer import infer
from collections import defaultdict
from tqdm import tqdm
```

```python
# current_dir = os.path.dirname(os.path.realpath(__file__))
current_dir = '.'
input_path = current_dir + '/input_data/'
print('input_path:', input_path, '\n')
```

```python
# Load spacy.
# nlp = spacy.load("en_core_web_trf")
nlp = spacy.load("en_core_web_sm")
```

```python
def predict(input_path, output_path):
    
    # Input path: ./input_data/
    # Paths for the folders of news topics.
    input_paths_topic = listdir_nohidden(input_path)
#     print('input_paths_topic:', len(input_paths_topic), '\n', input_paths_topic, '\n')
    
    for file_path in tqdm(input_paths_topic[:3]): # file_path <==> topic
#         print('file_path:', file_path)
        
        # file name
        fn = os.path.basename(file_path)
        print('file_name:', fn, '\n')
        
        # List of paths for the input files under each topic.
        input_file_paths = listdir_nohidden(file_path)
#         print('input_file_paths:', len(input_file_paths), '\n', input_file_paths, '\n')
        
        # Article files, format: {file_name: text}.
        file_dict_tokenized = defaultdict()
        for ifp in input_file_paths[:3]:
#             print('input_file_path:', ifp, '\n')
            aid = ifp.split('/')[-1][:-4]
#             print('fid:', fid, '\n')
#             for line in open(ifp, "r"):
#                 print(line, '\n\n')
#                 tokens = nlp(line)
#                 sentences = [str(text) for text in tokens.sents]
                
            with open(ifp, 'r') as article:
#                 print(article.readline(), '\n\n')
                text = article.readline()
#                 print('text:\n', text, '\n')
                tokens = nlp(text)
#                 print(type(tokens), '\n', tokens, '\n')
                sentences = [str(sent).strip() for sent in tokens.sents]
#                 print('sentences:\n', sentences, '\n\n')
        
                file_dict_tokenized[aid] = sentences
                
                
#             print('====================================================================')
            
        file_dict_tokenized = dict(file_dict_tokenized)
#         print(file_dict_tokenized.keys(), '\n')
        
        # Results of the event extraction.
        ee_results_maven = defaultdict(list)
        print("Start with event extraction.")
        start_time = time.time()
        for aid, sents in tqdm(file_dict_tokenized.items()):
            for sent in sents:
                if len(sent) > 0:
                    ee_results_maven[aid].append(infer(text=sent, schema='maven', task='EE')[0])
        end_time = time.time()
        print("End of event extraction.")
        print("--- %s seconds ---" % (end_time - start_time))
        ee_results_maven = dict(ee_results_maven)
        
        print(list(ee_results_maven.items())[0])
        
        # Write results to a file.
        with open(output_path + f'/{fn}_events.p', 'wb') as file:
            pickle.dump(ee_results_maven, file)
```

```python
predict(input_path=input_path, 
        output_path='./output_data/')
```
